<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Book в C#, который будет представлять информацию о
книгах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.



<h2 style="color:DodgerBlue">Реализация:</h2>

----

```
// Базовый класс Book
public class Book
{
    // Атрибуты
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }

    // Конструктор
    public Book(string title, string author, int yearOfPublication)
    {
        Title = title;
        Author = author;
        YearOfPublication = yearOfPublication;
    }

    // Методы
    public virtual string GetInfo()
    {
        return $"Книга: '{Title}', Автор: {Author}, Год издания: {YearOfPublication}";
    }

    public virtual void Read()
    {
        Console.WriteLine($"Чтение книги: '{Title}'");
    }

    public virtual void Borrow()
    {
        Console.WriteLine($"Книга '{Title}' выдана на чтение");
    }
}

// Производный класс Textbook (Учебник)
public class Textbook : Book
{
    // Дополнительный атрибут
    public string Subject { get; set; }

    // Конструктор
    public Textbook(string title, string author, int yearOfPublication, string subject) 
        : base(title, author, yearOfPublication)
    {
        Subject = subject;
    }

    // Переопределение метода Read()
    public override void Read()
    {
        Console.WriteLine($"Изучение учебника по предмету '{Subject}': '{Title}'");
    }

    // Переопределение метода GetInfo()
    public override string GetInfo()
    {
        return base.GetInfo() + $", Предмет: {Subject}";
    }
}

// Производный класс Fiction (Художественная литература)
public class Fiction : Book
{
    // Дополнительный атрибут
    public string Genre { get; set; }

    // Конструктор
    public Fiction(string title, string author, int yearOfPublication, string genre) 
        : base(title, author, yearOfPublication)
    {
        Genre = genre;
    }

    // Переопределение метода Borrow()
    public override void Borrow()
    {
        Console.WriteLine($"Художественная книга в жанре '{Genre}': '{Title}' выдана для удовольствия чтения");
    }

    // Переопределение метода GetInfo()
    public override string GetInfo()
    {
        return base.GetInfo() + $", Жанр: {Genre}";
    }
}

// Производный класс ScientificLiterature (Научная литература)
public class ScientificLiterature : Book
{
    // Дополнительный атрибут
    public string FieldOfScience { get; set; }

    // Конструктор
    public ScientificLiterature(string title, string author, int yearOfPublication, string fieldOfScience) 
        : base(title, author, yearOfPublication)
    {
        FieldOfScience = fieldOfScience;
    }

    // Переопределение метода GetInfo()
    public override string GetInfo()
    {
        return base.GetInfo() + $", Область науки: {FieldOfScience}";
    }

    // Переопределение метода Read()
    public override void Read()
    {
        Console.WriteLine($"Исследование научной работы в области '{FieldOfScience}': '{Title}'");
    }
}

// Демонстрация работы классов
public class LibraryDemo
{
    public static void Demonstrate()
    {
        Console.WriteLine("=== ДЕМОНСТРАЦИЯ РАБОТЫ БИБЛИОТЕКИ ===\n");

        // Создание объектов разных типов книг
        Book[] books = new Book[]
        {
            new Book("Основы программирования", "Иван Петров", 2020),
            new Textbook("Математика для начинающих", "Анна Сидорова", 2019, "Математика"),
            new Fiction("Преступление и наказание", "Федор Достоевский", 1866, "Роман"),
            new ScientificLiterature("Теория относительности", "Альберт Эйнштейн", 1916, "Физика"),
            new Fiction("1984", "Джордж Оруэлл", 1949, "Антиутопия")
        };

        // Демонстрация полиморфизма
        foreach (var book in books)
        {
            Console.WriteLine(book.GetInfo());
            book.Read();
            book.Borrow();
            Console.WriteLine();
        }

        // Дополнительная демонстрация специфического поведения
        Console.WriteLine("=== СПЕЦИФИЧЕСКОЕ ПОВЕДЕНИЕ ===");
        
        Textbook mathTextbook = new Textbook("Высшая математика", "Сергей Иванов", 2018, "Математика");
        mathTextbook.Read(); // Специфическое поведение учебника
        
        Fiction novel = new Fiction("Война и мир", "Лев Толстой", 1869, "Эпопея");
        novel.Borrow(); // Специфическое поведение художественной литературы
        
        ScientificLiterature physicsBook = new ScientificLiterature("Квантовая механика", "Нильс Бор", 1925, "Физика");
        Console.WriteLine(physicsBook.GetInfo()); // Специфическое информационное описание
    }
}

// Запуск демонстрации
LibraryDemo.Demonstrate();
```
